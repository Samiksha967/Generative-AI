```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
np.random.seed(0)

data1 = np.random.normal(0, 1, (100, 2))
data2 = np.random.normal(5, 1, (100, 2))

X = np.vstack((data1, data2))
```

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(X)

print("GMM trained successfully ✅")
```

```python
new_data, _ = gmm.sample(200)
```

```python
plt.scatter(X[:,0], X[:,1], label="Original Data")
plt.scatter(new_data[:,0], new_data[:,1], label="Generated Data")
plt.legend()
plt.title("Gaussian Mixture Model")
plt.show()
```

Build a simple text generator using Markov Chains. 

```python
import random
```

```python
text = """
artificial intelligence is changing the world
intelligence is the ability to learn
machine learning is a part of artificial intelligence
"""
```

```python
def build_markov_chain(text):
    words = text.split()
    chain = {}

    for i in range(len(words)-1):
        word = words[i]
        next_word = words[i+1]

        if word not in chain:
            chain[word] = []
        chain[word].append(next_word)

    return chain
```

```python
def generate_text(chain, start_word, length=15):
    word = start_word
    result = [word]

    for _ in range(length):
        if word not in chain:
            break
        word = random.choice(chain[word])
        result.append(word)

    return " ".join(result)
```

```python
markov_chain = build_markov_chain(text)

generated_text = generate_text(markov_chain, "artificial", 20)
print("Generated Text:")
print(generated_text)
```
