Import Libraries



```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt
```

Load Stable Diffusion Model

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

Write Creative Prompt

```python
prompt = """
A futuristic cyberpunk city at sunset,
neon lights reflecting on wet streets,
flying cars, ultra detailed digital art,
8k resolution, cinematic lighting
"""
```

Generate Artwork

```python
image = pipe(prompt).images[0]

plt.imshow(image)
plt.axis("off")
```

Save the Generated Artwork

```python
image.save("stable_diffusion_art.png")

print("Artwork saved successfully!")
```

Generate Multiple Creative Images

```python
prompts = [
    "A fantasy dragon flying over a medieval castle, digital art",
    "Astronaut riding a horse on Mars, surreal art",
    "Ancient temple hidden inside a glowing jungle, cinematic lighting"
]

for p in prompts:

    img = pipe(p).images[0]

    plt.figure()
    plt.imshow(img)
    plt.axis("off")
```

Higher Quality Image (Optional)

```python
image = pipe(
    prompt,
    num_inference_steps=50,
    guidance_scale=8
).images[0]

plt.imshow(image)
plt.axis("off")
```
