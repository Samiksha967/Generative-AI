Helper function to simulate model responses without API

```python

def simulate_model(prompt_type, task):
    """
    Simulates model response for different prompt engineering approaches.
    """
    if prompt_type == "interview":
        return """Step 1: Convert 1 hour 30 minutes to hours -> 1.5 hours
Step 2: Formula for average speed = Distance / Time
Step 3: Average speed = 60 / 1.5 = 40 km/h
Answer: 40 km/h"""
    
    elif prompt_type == "cot":
        return """Let's think step by step:
1. Convert time: 1 hour 30 minutes = 1.5 hours
2. Average speed = Distance / Time = 60 / 1.5
Answer: 40 km/h"""
    
    elif prompt_type == "tot":
        return """Option 1: Day1: Louvre + Eiffel Tower, Day2: Montmartre + Seine Cruise, Cost = $300
Option 2: Day1: Versailles + Louvre, Day2: Notre Dame + Eiffel Tower, Cost = $350
Best Option: Option 1 (balanced sightseeing, cost-effective)"""
    
    elif prompt_type == "zero_shot":
        return "Bonjour, comment ça va ?"
    
    elif prompt_type == "few_shot":
        return "Comment ça va ?"
    
    else:
        return "Unknown prompt type"
```

Interview Approach Example

```python
# Stepwise guided prompts to solve a problem
interview_response = simulate_model("interview", "Calculate train speed")
print("Interview Approach Output:\n", interview_response)
```

Chain-of-Thought (CoT) Example

```python
# Model explains reasoning step by step
cot_response = simulate_model("cot", "Calculate train speed")
print("Chain-of-Thought Output:\n", cot_response)
```

Tree-of-Thought (ToT) Example

```python
# Model explores multiple solution paths and chooses the best
tot_response = simulate_model("tot", "Plan Paris trip")
print("Tree-of-Thought Output:\n", tot_response)
```

Zero-Shot Prompting Example

```python
# Model is asked to perform a task without examples
zero_shot_response = simulate_model("zero_shot", "Translate to French")
print("Zero-Shot Output:\n", zero_shot_response)
```

Few-Shot Prompting Example

```python
# Model is given few examples before performing the task
few_shot_response = simulate_model("few_shot", "Translate to French")
print("Few-Shot Output:\n", few_shot_response)
```
